```python
import pandas as pd
import numpy as np

df = pd.read_csv("../train.csv")

print("Dataset Loaded")
```

```python
# House Age Feature

df['HouseAge'] = (
    df['YrSold'] - df['YearBuilt']
)

df[['HouseAge']].head()
```

```python
# Total Square Feet

df['TotalSF'] = (
    df['TotalBsmtSF']
    + df['1stFlrSF']
    + df['2ndFlrSF']
)

df[['TotalSF']].head()
```

```python
# Total Bathrooms

df['TotalBath'] = (
    df['FullBath']
    + (0.5 * df['HalfBath'])
    + df['BsmtFullBath']
    + (0.5 * df['BsmtHalfBath'])
)

df[['TotalBath']].head()
```

```python
# Log Transformation

df['SalePrice'] = np.log1p(
    df['SalePrice']
)

df[['SalePrice']].head()
```
